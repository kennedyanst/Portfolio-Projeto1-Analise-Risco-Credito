# Uso de gráfico de dispersão da biblioteca Plotly para detectar possíveis outliers

## Obs: Utilizei o Microsot Power BI para uma visualização mais refinada dos gráficos. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv("credit_data.csv")
```

```python
# Identificando valores vázios 
base.isnull().sum()
```

```python
# Dropando os valores vázios
base.dropna(inplace=True)
```

```python
base.isnull().sum()
```

# Uso de gráficos de dispersão nos parâmetros:

* Renda x Idade
* Renda x Dívida
* Idade x Dívida

## Objetivo: Analisar possíveis dados incorretos na base de dados.

# Idade x Dívida | Renda x Idade | Renda x Idade 

```python
fig, eixos = plt.subplots(1, 3, figsize=(12,5))

eixos[0].scatter(x=base["age"], y=base["loan"])
eixos[0].set_title("Idade x Dívida")

eixos[1].scatter(x=base["income"], y=base["age"])
eixos[1].set_title("Renda x Idade")

eixos[2].scatter(x=base["income"], y=base["loan"])
eixos[2].set_title("Renda x Dívida");
```

Nota-se os valores negativos da idade na base de dados, que serão retirados na fase de pré-processamento e limpeza dos dados.
